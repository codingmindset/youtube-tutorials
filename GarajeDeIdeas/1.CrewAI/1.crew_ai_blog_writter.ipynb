```python
from crewai import Agent, Task, Process, Crew
from crewai_tools import SerperDevTool
from langchain.agents import load_tools
```

```python
search_tool = SerperDevTool()
```

```python
human_tools = load_tools(["human"])
```

```python
researcher = Agent(
    role="Senior Researcher",
    goal="Discover and investigate leading-edge initiatives and enterprises within the AI and machine learning sectors for the year 2024. Find one article and pass it to the SEO Writter Specialist agent",
    backstory="""As a seasoned strategist, you excel in identifying nascent trends and noteworthy entities in AI, technology, and machine learning. Your expertise includes transforming web-scraped information 
    into comprehensive analyses, listing the most promising projects and firms in the AI/ML domain. This report relies exclusively on data obtained from internet sources.""",
    verbose=True,
    allow_delegation=False,
    tools=[search_tool],
)
```

```python
seo_writer_junior = Agent(
  role='SEO Writter Specialist',
  goal='Based on the search results, write a blog post with SEO best practices',
  backstory='An expert in SEO and content writing, you know how to create engaging content that ranks well on search engines.',
  tools=[search_tool] + human_tools,
  verbose=True,
  allow_delegation=False
)
```

```python
seo_writer_senior = Agent(
    role="Senior SEO Content Writer",
    goal="Optimize blog posts to enhance their search engine visibility and engagement. Ensure that content is SEO-friendly, impactful, and reader-focused",
    backstory="""You are an adept SEO specialist with a proven track record of elevating online content. Your expertise lies in integrating targeted keywords 
    seamlessly while maintaining the narrative's flow and readability. You are skilled at adjusting the structure and metadata of articles to maximize their 
    search engine rankings without compromising content quality. With your guidance, blog posts not only attract more views but also maintain a natural tone 
    that resonates with readers.""",
    verbose=True,
    allow_delegation=False,
)
```

```python
task_research = Task(
    description="""Research and compile a detailed list of the most promising and innovative AI projects and tools currently available. 
    Use only scraped data from trusted internet sources to gather this information. The final report should be in a bullet-point format, highlighting 5-10 new AI projects and tools.
    Each bullet point must include three sentences detailing one specific AI company, product, model, or tool, including its relevance, innovation, and potential impact in the field.""",
    expected_output='Text report in bullet-point format with detailed descriptions.',
    tools=[search_tool],  # Assuming the seo_writer has an appropriate research tool
    agent=researcher,
)
```

```python
task_write_seo_blog = Task(
    description="""Write an SEO-optimized blog post based on the research conducted on AI tools and projects. The blog should be at least 10 paragraphs long, with a compelling and catchy headline. 
    The style should be engaging yet easy to understand, using layman's terms where necessary to ensure it is accessible to the general public. Each mention of an AI project or tool should be bolded. Start each new paragraph on a new line, without labeling paragraphs numerically.""",
    expected_output='Markdown formatted text of the blog post, including headlines, bolded project/tool names, and links.',
    tools=[search_tool],  # Assuming the seo_writer has a writing tool suitable for this task
    agent=seo_writer_junior,
)
```

```python
task_optimize_seo_blog = Task(
    description="""Review and optimize the drafted blog post to enhance its SEO effectiveness. Ensure that the blog includes proper keyword density, strategic use of meta tags, and is structured for easy readability and engagement. 
    Adjust any sections that require refinement to boost search engine rankings and reader engagement, while maintaining technical accuracy and insight.""",
    expected_output='A fully optimized SEO-friendly blog post ready for publication.',
    agent=seo_writer_senior,
)
```

```python
crew = Crew(
    agents=[researcher, seo_writer_junior, seo_writer_senior],
    tasks=[task_research, task_write_seo_blog, task_optimize_seo_blog],
    process=Process.sequential,
    memory=True,
    cache=True,
    share_crew=False
)
```

```python
# Initiate the task execution process
# result = crew.kickoff(inputs={'topic': 'latest AI tools and projects'})
result = crew.kickoff()
print(result)
```
